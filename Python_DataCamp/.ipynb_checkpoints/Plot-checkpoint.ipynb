# 1. array 2-D plot

## 1.1 2D smooth function map

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.linspace(-2, 2, 10)
v = np.linspace(-1, 1, 10)
X,Y = np.meshgrid(u, v) 
Z = X**2/25 + Y**2/4
```

```python
plt.pcolor(Z, cmap='autumn')   # pcolor: pseudocolor; cmap: color map
plt.show()
```

## 1.2 Distributions - 2D histogram

```python
x = np.array([1,3,4,5,6,4,3,8,5,6,4,4,9,6,3,5])
y = np.array([1,2,9,5,6,8,5,4,5,4,6,4,9,7,3,9])
```

```python
plt.hexbin(x, y, gridsize=(15,10),extent=(2, 8, 2, 8))    
```

```python
plt.hist2d(x, y, bins=(10, 20), range=((2, 8), (2, 8)))  #((xmin,xmax),(ymin,ymax))
```

# 2. Seaborn

```python
auto=pd.read_csv('auto.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# linear regression
sns.lmplot(x='weight', y='hp', data=auto)
```

```python
sns.residplot(x='hp', y='mpg', data=auto, color='green')
```

```python
sns.regplot(x='weight', y='mpg', data=auto, color='blue', label='First Order')
```

```python
# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None, color='blue', label='First Order')

# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None, color='green', order=2, label='Second Order')

# Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()
```
